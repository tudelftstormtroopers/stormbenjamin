```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial import KDTree
import numpy as np

#V1 and V2 are quality values. when they are too high (+/-5), the elevation (z) is unreliable.

# Function to read and parse the data from a file
def read_gps_data(file_path):
    # Read the tab-separated file, assuming no header
    df = pd.read_csv(file_path, sep='\t', header=None, 
                     names=['name', 'date', 'time', 'x', 'y', 'z', 'v1', 'v2'],
                     parse_dates={'datetime': ['date', 'time']}, dayfirst=True)
    
    # Convert x, y, z to float if not already
    df['x'] = pd.to_numeric(df['x'], errors='coerce')
    df['y'] = pd.to_numeric(df['y'], errors='coerce')
    df['z'] = pd.to_numeric(df['z'], errors='coerce')
    
    # Drop any rows with NaN in coordinates or z
    df = df.dropna(subset=['x', 'y', 'z'])
    
    return df

# Load the two datasets
file1 = 'STORMTROOP 12 5_data of 24Oct only.txt'
file2 = 'STORMTROOP 12 5_data of 30Oct only.txt'

df1 = read_gps_data(file1)
df2 = read_gps_data(file2)


df1_filtered = df1[(df1['x'] < 80000) & (df1['y'] > 450000)]
df2_filtered = df2[(df2['x'] < 80000) & (df2['y'] > 450000)]

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df1_filtered["x"], df1_filtered["y"], df1_filtered["z"], color='green', label='Pre-storm')
ax.scatter(df2_filtered["x"], df2_filtered["y"], df2_filtered["z"], color='blue', label='Post-storm')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot of GPS Data Pre- and Post-Storm')
ax.legend()
plt.show()
```
